```python
def plotowanie(x, y, n_ch, Fs, titlex='Czas [s]', titley='Amplituda [µV]',xmin=0, xmax=0, title='Sygnał z kanału '):
    if xmin == 0:
        xmin = x[0]
    if xmax == 0:
        xmax=xmin+len(y)/Fs
    if type(n_ch) is str:
        plt.plot(x, y)
        plt.xlabel(titlex)
        plt.ylabel(titley)
        plt.title(title+str(n_ch)) 
        plt.tight_layout()
        plt.xlim(xmin, xmax)
    else:
        for ch in range(n_ch):
            plt.subplot(n_ch,1,ch+1)
            plt.plot(x, y[:,ch])
            plt.xlabel(titlex)
            plt.ylabel(titley)
            plt.title(title+str(ch+1)) 
            plt.tight_layout()
            plt.xlim(xmin, xmax)
    plt.show()
```
